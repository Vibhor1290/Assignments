```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

X, y = make_moons(n_samples=1200, noise=0.3, random_state=35)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

db_model = DBSCAN(eps=0.3, min_samples=5)
clusters = db_model.fit_predict(X_scaled)

silhouette = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score: {silhouette:.3f}")

plt.figure(figsize=(10, 6))
unique_clusters = np.unique(clusters)
colors = plt.cm.get_cmap("viridis", len(unique_clusters))

for cluster in unique_clusters:
    if cluster == -1:
        # Noise points
        color = 'k'
        marker = 'x'
    else:
        color = colors(cluster)
        marker = 'o'
    cluster_points = X_scaled[clusters == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=color, marker=marker, label=f'Cluster {cluster}')
    
plt.title('DBSCAN Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
