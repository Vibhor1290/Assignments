Q7. What is the make_circles package in scikit-learn used for?

```python
from sklearn.datasets import make_circles
```

```python
X,Y = make_circles(n_samples=2000,random_state=42,noise=0.3,shuffle=True)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.8,min_samples=5)
```

```python
pred = dbscan.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis', s=40)
plt.title('Generated Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=pred, cmap='viridis', s=40)
plt.title('DBSCAN Cluster')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Q9. How can local outliers be detected using the Local Outlier Factor (LOF) algorithm?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.neighbors import LocalOutlierFactor
```

```python
X,Y = make_circles(n_samples=2000,noise=0.3,factor=0.7)
```

```python
LOF = LocalOutlierFactor(n_neighbors=10,contamination=0.2)
```

```python
y_pred = LOF.fit_predict(X)
```

```python
outlier_score = -LOF.negative_outlier_factor_
```

```python
outlier_score
```

```python
plt.figure(figsize=(6, 6))
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.scatter(X[:, 0], X[:, 1], color='k', s=3, label='Data points')
```

```python
radius = (X.max(axis=0) - X.min(axis=0)).max() / 2.0
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], 
            color='red', s=20*radius, edgecolors='k', label='Outliers')

plt.title('Local Outlier Factor (LOF)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

Q10. How can global outliers be detected using the Isolation Forest algorithm?

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles
from sklearn.ensemble import IsolationForest

X, y = make_circles(n_samples=3000, noise=0.3, factor=0.2, random_state=50)

IsoForest = IsolationForest(random_state=42, contamination=0.2)
IsoForest.fit(X)


y_pred = IsoForest.predict(X)
outliers = np.where(y_pred == -1)

plt.figure(figsize=(8, 8))

plt.scatter(X[:, 0], X[:, 1], color='k', s=20, label='Data points')

# Plot outliers
plt.scatter(X[outliers, 0], X[outliers, 1], color='red', s=50, edgecolors='k', label='Outliers')
plt.title('Isolation Forest')
plt.xlabel('Feature A')
plt.ylabel('Feature B')
plt.legend()
plt.show()
```
