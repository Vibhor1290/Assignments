Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
from scipy.stats import norm

sample = 500
smokers_count = 60

sample_proportion = smokers_count / sample

standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample)

confidence_level = 0.95

z_value = norm.ppf((1 + confidence_level) / 2)

marginal_error = z_value * standard_error

lower_bound = sample_proportion - marginal_error
upper_bound = sample_proportion + marginal_error

print([lower_bound,upper_bound])
```

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

```python
sample_size = 1000
coke_people = 520

sample_proportion = coke_people / sample_size

standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

confidence_level = 0.99

z_score = norm.ppf((1 + confidence_level) / 2)

margin = z_score * standard_error

lower_bound = sample_proportion - margin
upper_bound = sample_proportion + margin

# Print the confidence interval
print([lower_bound,upper_bound])
```

Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

```python
# Null hypo: The coin is fair

# Alt hypo: The coin is biased towards tail
```

```python
import scipy.stats as sc

total_trials = 100
observed_tails = 45

expected_heads = total_trials/ 2
expected_tails = total_trials/ 2

observed_data = np.array([observed_tails, total_trials - observed_tails])

expected_data = np.array([expected_tails, expected_heads])

chi2, p_value = sc.chisquare(observed_data, expected_data)

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Failed to reject null hypo")
```
